getApeBoardPriceMomentum(path) takes the local path to your exported ApeBoard data, and returns a weighted index on your portfolio's peformance over the past week based on the amount you own of each asset and it's performance over the past week. Data is taken from coinlist's api.

```python

def getApeBoardPricemomentum(path):
    df = pd.read_csv(path)
    URL1 = "https://api.coinstats.app/public/v1/coins?skip=0&limit=100&currency=USD"
    page1 = requests.get(URL1)
    jsonpage = json.loads(page1.text)
    
    volume_list= []
    price_list = []
    name_list = []
    symbol_list = []
    priceChange1w_list = []
    
    for i in range(len(jsonpage['coins'])):
        if jsonpage['coins'][i]['symbol'] in str(df['symbol']):
            volume_list.append(jsonpage['coins'][i]["volume"])
            price_list.append(jsonpage['coins'][i]["price"])
            name_list.append(jsonpage['coins'][i]['name']) 
            priceChange1w_list.append(jsonpage['coins'][i]["priceChange1w"])
            symbol_list.append(jsonpage['coins'][i]["symbol"])
    d = {
    "volume": volume_list,
    "price": price_list,
    "name": name_list,
    "price_change_1w": priceChange1w_list,
    "symbol": symbol_list}
    pdf = pd.DataFrame(d)
    sortedpdf = pdf
    sorteddf = df.drop(0)
    
    #calculate the total value of portfolio
    # total portfolio price change = pc1*(val1/total) + ....
    total = sorteddf['balance'].sum()
    price_movement_1w = 0
    
    for i in symbol_list:
        dfloc = sorteddf.loc[sorteddf['symbol'] == i]
        pdfloc = sortedpdf.loc[sortedpdf['symbol'] == i]
        bal = dfloc['balance']
        chng = pdfloc['price_change_1w']
        price_movement_1w = price_movement_1w + (chng.item() * (bal.item()/total))
        
    return "Your Apeboard portfolio's weighted price movement index for the past week is " + str(price_movement_1w) + "%"
    
```
